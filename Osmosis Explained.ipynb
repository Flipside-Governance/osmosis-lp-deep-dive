```python
## Import Libraries 
import matplotlib.pyplot as plt
from matplotlib import cm 
from matplotlib.ticker import LinearLocator
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
## Simulated Two-Token Liquidity Pool - Set Variables
y = range(5000, 15000, 100)
k = 45000000
x = []
for i in range(0, len(y)): 
    ##print(i)
    x.append(k/y[i])
```

```python
# Simulated Two-Token Liquidity Pool - Create Plot
plt.plot(x, y, label='Constant Product')
plt.plot(4500, 10000, marker="o", markersize=6, markeredgecolor="red",
markerfacecolor="red", label='Current Price')
plt.xlabel("OSMO")
plt.ylabel("Evmos")
plt.legend(loc="upper right")
```

```python
## Real Two-Token Liquidity Pool - Set Variables
y = range(3046513, 5046513, 10000)
k = 4046513*1727555
x = []
for i in range(0, len(y)): 
    ##print(i)
    x.append(k/y[i])
```

```python
# Real Two-Token Liquidity Pool - Create Plot
plt.plot(x, y, label='Constant Product')
plt.plot(1727555, 4046513, marker="o", markersize=6, markeredgecolor="red",
markerfacecolor="red", label='Current Price')
plt.xlabel("OSMO")
plt.ylabel("Evmos")
plt.legend(loc="upper right")
```

```python
## Simulated Three-Token Liquidity Pool - Set Variables
y = np.arange(1500, 2500, 100) # JKL Token
x = np.arange(500, 1500, 100) # FET Token
k = 1000*2000*20000
x, y = np.meshgrid(x, y)
z = (k/(y*x)) #BCNA Token
```

```python
## Simulated Three-Token Liquidity Pool
fig = plt.figure(1)
ax = Axes3D(fig, computed_zorder=False)

surf = ax.plot_surface(x, y, z, cmap=cm.plasma, linewidth=0, antialiased=False, label='Constant Product')
ax.scatter(1000, 2000, 20000, s=20, marker='o', color = "k", cmap=None, label='Current Price')

ax.set_xlabel('FET')
ax.set_ylabel('JKL')
ax.yaxis.labelpad = 7
ax.set_zlabel('BCNA')
ax.zaxis.labelpad = 12
plt.tight_layout()
plt.show()
```

```python
## Real Three-Token Liquidity Pool - Set Variables
y2 = np.arange(1794, 2794, 100) # JKL Token
x2 = np.arange(971, 1971, 100) # FET Token
k2 = 1471*2294*31598
x2, y2 = np.meshgrid(x2, y2)
z2 = (k2/(y2*x2)) #BCNA Token
```
